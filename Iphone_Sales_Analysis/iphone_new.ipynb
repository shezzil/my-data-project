```python
import pandas as pd
```

```python
df = pd.read_csv('C:/My Files/Courses/Python/Project_File/iphone.csv')
df.head()
```

```python
# 1 - The column names have spaces . rename the column names to have underscore '_' instead of space 
#  (try to do in one go instead of specifying each column nam in rename method)

df.columns = df.columns.str.replace(' ', "_")
df
```

```python
df.isnull().sum()
```

```python
# 2- star rating for some of the models is missing in the dataset. fill those missing values with the average 
#    rating all the models.

mean_rating = df['Star_Rating'].fillna(df['Star_Rating'].mean())
mean_rating
```

```python
# 3- Now instead of filling missing values with avg rating of full dataset , fill with avg rating based on RAM.
# example: if rating for a 2 gb iphone is missing then take average of all other 2 gb phones rating and fill that value.

def missing_value(row):
    df_avg_rating = df.groupby('Ram')['Star_Rating'].mean()
    if pd.isna(row['Star_Rating']):
        return df_avg_rating.get(row['Ram'], None)
    return row['Star_Rating']

df['Star_Rating'] = df.apply(missing_value, axis=1)
df['Star_Rating']
```

```python
# 4- create a new column in the dataframe "Discount_Percentage" based on MRP and sale value
df['Discount_Percentage'] = ((df['Mrp'] - df['Sale_Price'])/df['Mrp'])*100
df.head()
```

```python
#5- which model has highest percent discount ?

df['discount'] = (df['Mrp'] - df['Sale_Price'])/df['Mrp']*100
sorted_df = df.sort_values(by='discount', ascending=False)
sorted_df.iloc[1]['Product_Name']
```

```python
# 6- find total no of models  each space configuration (128 GB , 64 GB etc)

model_counts = df['Ram'].value_counts()
model_counts.columns = ['Space_Configuration', 'Total_Models']
print(model_counts)
```

```python
# 7- find total number of models for each color 

def extract_color(product_name):
    parts = product_name.split('(')
    if len(parts) > 1:
        color_part = parts[1].split(',')
        actual_color = color_part[0].strip()
        return actual_color
    return None

df['color'] = df['Product_Name'].apply(extract_color)
color_count = df['color'].value_counts().reset_index
color_count

```

```python
# 8- find total number of models by iphone version 

def fetch_model(product_name):
    name = product_name.split('(')
    model_name = name[0].strip()
    return model_name

df['models'] = df['Product _Name'].apply(fetch_model)
df['models'].value_counts()
```

```python
# 9- list top 5 models having highest no of reviews 

top_5 = df.sort_values(by='Number _Of _Reviews', ascending=False)
top_5.head(5)
```

```python
# 10 - what is the price diffrence between highest price and lowest price iphone (based on mrp)

highest_price = df['Mrp'].max()
lowest_price = df['Mrp'].min()

PriceDifferenceHighestLowest_Mrp = highest_price - lowest_price
PriceDifferenceHighestLowest_Mrp
```

```python
# 11-find total no of reviews for iphone 11 and iphone 12 category . Output should have only 2 rows (for 11 and 12).

iphone_11_reviews = df[df['Product _Name'].str.contains('iPhone 11', case=False)]
iphone_12_reviews = df[df['Product _Name'].str.contains('iphone 12', case = False)]

total_reviews_11 = iphone_11_reviews['Number _Of _Reviews'].sum()
total_reviews_12 = iphone_12_reviews['Number _Of _Reviews'].sum()

final_result = pd.DataFrame({
    'iphone_category' : ['iphone 11', 'iphone 12'],
    'Total_reviews' : [total_reviews_11, total_reviews_12]
})

final_result
```

```python
# 12- which iphone has 3rd highest MRP

sorting = df.sort_values(by='Mrp', ascending=False).reset_index(drop = True)
third_highest_mrp_iphone = sorting.loc[2]

print(third_highest_mrp_iphone)
```
